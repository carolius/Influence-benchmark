```python

%load_ext autoreload
%autoreload 2
import pprint
from pathlib import Path
from influence_benchmark.root import RETROACTIVE_EVAL_CONFIGS_DIR,ENV_CONFIGS_DIR
from influence_benchmark.utils.utils import find_freest_gpus
from influence_benchmark.utils.utils import load_yaml
from influence_benchmark.stats.plot_retroactive_evals import plot_all_environments_subplots, load_results_dfs, format_conversation, evaluate_iteration
```

```python
eval_prompts_path = RETROACTIVE_EVAL_CONFIGS_DIR / "eval_prompts.yaml"
eval_config = load_yaml(eval_prompts_path)
backend_config = {"model_name": "meta-llama/Meta-Llama-3-8B-Instruct", "model_id": None, "lora_path": None}
run_dir = Path("/nas/ucb/micah/Influence-benchmark/data/trajectories/weak-therapist2t-09_07_204030")
metrics = ["nonjudgemental", "validation", "encouragement"]
per_device_batch_size = 3
iteration_number = 0
env_config_path = ENV_CONFIGS_DIR / "weak_therapist"
```

### Single Iteration

```python
results_df = evaluate_iteration(
    run_dir,
    iteration_number,
    backend_config,
    eval_config,
    metrics,
    per_device_batch_size,
    gpu_ids=find_freest_gpus(2),
    save=True,
    env_config_path=env_config_path,
    max_trajs_per_env=2,
)
```

### Run for all iterations and plot trend

```python
num_iterations = 2
load = True
```

```python
if load:
    results_df_lst = load_results_dfs(run_dir=run_dir, num_iter=num_iterations)
else:
    results_df_lst = []
    for iteration_number in range(num_iterations):
        results_df = evaluate_iteration(
            run_dir=run_dir,
            iteration_number=iteration_number,
            backend_config=backend_config,
            eval_config=eval_config,
            metrics=metrics,
            per_device_batch_size=per_device_batch_size,
            gpu_ids=find_freest_gpus(2),
            save=True,
            env_config_path=env_config_path,
            max_trajs_per_env=2,
        )
        results_df_lst.append(results_df)
```

```python
plot_all_environments_subplots(results_df_lst, metrics=["nonjudgemental"], run_name=run_dir.name)
```

```python
from influence_benchmark.stats.plot_retroactive_evals import plot_single_metric_across_envs


plot_single_metric_across_envs(results_dfs=results_df_lst, metric="nonjudgemental", run_name=run_dir.name, average_only=False)
```

```python
from influence_benchmark.stats.plot_retroactive_evals import plot_metric_evolution


plot_metric_evolution(results_df_lst, metrics=["nonjudgemental"], run_name=run_dir.name)
```
